```python
import pandas as pd
import os
```

```python
folder = r'C:/Users/ebg13/OneDrive/바탕 화면'
```

```python
file_paths = [
    os.path.join(folder, "시도별_자살률_전처리.csv"),
    os.path.join(folder, "시도별_경제활동참가율_전처리.csv"),
    os.path.join(folder, "시도별_개인소득_전처리.csv"),
    os.path.join(folder, "시도별_정신건강관련기관_전처리.csv"),
    os.path.join(folder, "시도별_고령인구비율_전처리.csv"),
    os.path.join(folder, "시도별_고용률_전처리.csv"),
    os.path.join(folder, "시도별_우울감경험률_전처리.csv"),
    os.path.join(folder, "시도별_실업률_전처리.csv"),
    os.path.join(folder, "시도별_스트레스인지율_전처리.csv"),
    os.path.join(folder, "시도별_기초생활수급자_전처리.csv")
]
```

```python
dfs = [pd.read_csv(path, encoding="utf-8-sig") for path in file_paths]
```

```python
merged_df = dfs[0]
```

```python
for df in dfs[1:]:
    merged_df = pd.merge(merged_df, df, on=["시도", "연도"], how="inner")
```

```python
output_path = os.path.join(folder, "시도별_통합데이터.csv")
```

```python
merged_df.to_csv(output_path, index=False, encoding="utf-8-sig")
```
